```python
## Step 1: Create a Sample DataFrame (15 rows × 6 columns)
import pandas as pd
import numpy as np

# Create sample DataFrame
data = {
    'EmployeeID': range(1, 16),
    'YearsExperience': [1, 2, 3, 5, np.nan, 7, 8, 2, 4, 9, 10, 6, np.nan, 3, 4],
    'EducationLevel': ['Bachelors', 'Masters', 'Bachelors', 'Masters', 'PhD',
                       np.nan, 'Bachelors', 'Diploma', 'Masters', 'PhD',
                       'Diploma', 'Masters', 'PhD', np.nan, 'Bachelors'],
    'SkillsScore': [78, 85, np.nan, 92, 88, 70, np.nan, 80, 75, 95, 65, np.nan, 90, 82, 76],
    'Department': ['IT', 'HR', 'Finance', 'Finance', 'IT', 'HR', np.nan,
                   'IT', 'Finance', 'HR', 'Finance', 'IT', 'HR', 'HR', np.nan],
    'Hired': [1, 0, 1, 1, np.nan, 0, 1, 0, 1, 1, 0, 1, np.nan, 0, 1]
}

df = pd.DataFrame(data)

df
```

## Data Preparation

## Dealing with null values
- Real inputs 
- drop nulls 
- filling null values mean or median for numerical data
- mean gets affected but ouliers 
- median (numerial data)
- categorical mode 

```python
## null values
df.isnull().sum()              # Count missing values
#df[df['SkillsScore'].isnull()] # Show rows where SkillsScore is missing
```

```python
df_filled = df.fillna({
    'YearsExperience': df['YearsExperience'].mean(),
    'SkillsScore': df['SkillsScore'].median(),
    'EducationLevel': 'Masters',
    'Department': 'Unknown',
    'Hired': df['Hired'].mode()[0]
})

#df_dropped = df.dropna()       # remove all rows with any missing value
```

```python
## column 
df['SkillsScore']
df[['YearsExperience', 'SkillsScore']]
```

```python
## Filter functions
df[df['YearsExperience'] > 5]
```

```python
df[(df['Department'] == 'IT') & (df['Hired'] == 1)]
```

```python
df
```

```python
## Modify or add columns
df['ScoreCategory'] = np.where(df['SkillsScore'] > 80, 'High', 'Low')
```

```python
df
```

```python
## remove Columns
df.drop(columns=['ScoreCategory'], inplace=True)
```

```python
df
```

```python
##Grouping & Aggregation
df.groupby('Department')['SkillsScore'].mean()
df.groupby('EducationLevel')['Hired'].sum()
```

```python
## Sorting 
df.sort_values('YearsExperience')
```

```python

df.sort_values(['Department', 'SkillsScore'], ascending=[True, False])
```

```python
## Converting Datatypes
df['EducationLevel'] = df['EducationLevel'].astype('category')
df['Hired'] = df['Hired'].astype('float')
```

```python
## Saving clean file
df.to_csv("cleaned_data.csv", index=False)
```
