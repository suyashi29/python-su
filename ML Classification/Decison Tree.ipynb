## 1. What Is a Decision Tree?

A Decision Tree Algorithm is a supervised learning method used for classification and regression tasks. It works by splitting data into branches based on feature values, forming a tree-like structure where each internal node represents a decision on a feature, and each leaf node represents an outcome (class label or predicted value).


### How It Works ? 

Root Node: Represents the entire dataset.
Splitting: The algorithm chooses the best feature to split the data based on a criterion (e.g., Gini Index, Entropy for classification; Variance Reduction for regression).
Branches: Each split creates branches for possible feature values.
Leaf Nodes: Final nodes that represent predictions.


### Key Concepts

#### Impurity Measures:

- Gini Index: Measures how often a randomly chosen element would be incorrectly classified.
- Entropy: Measures information gain (used in ID3 algorithm).
- Information Gain: How much uncertainty is reduced by splittin
- Entropy = − Σ pᵢ log₂(pᵢ)
```
Higher entropy = more disorder.


---
### Information Gain
Shows how much entropy is reduced:
```
Information Gain = Entropy(parent) − Σ weighted child entropy
`

### Step 1: Entropy of target
Yes=5, No=3
```
Entropy = - (5/8 log₂ 5/8 + 3/8 log₂ 3/8) = 0.954
```


### Step 2: Entropy after splitting by Weather
- Sunny (3 samples → 2 No, 1 Yes): Entropy = 0.918
- Overcast (2 samples → all Yes): Entropy = 0
- Rain (3 samples → 2 Yes, 1 No): Entropy = 0.918


Weighted entropy:
```
(3/8)*0.918 + (2/8)*0 + (3/8)*0.918 = 0.688
```
Information Gain:
```
IG = 0.954 – 0.688 = 0.266
```
**Overcast is pure → best initial split**

### Simple Example to Understand Decision Tree

````python
from sklearn.tree import DecisionTreeClassifier

# --- Step 1: Create a tiny dataset ---
# Weather data to decide whether to play outside
# 0 = No, 1 = Yes
X = [
    [85, 85],   # Hot, High humidity
    [80, 90],   # Hot, Very high humidity
    [72, 95],   # Mild, Very high humidity
    [69, 70],   # Cool, Normal humidity
    [75, 80],   # Mild, High humidity
]

# Target variable (play or not)
y = [0, 0, 1, 1, 1]

# --- Step 2: Build Decision Tree ---
model = DecisionTreeClassifier()
model.fit(X, y)

# --- Step 3: Predict ---
print("Prediction for a cool day with humidity 60:")
print(model.predict([[65, 60]]))
````
