### **1. What are Impurity Measures?**

Impurity measures tell us **how mixed** a node is.

* A node with **only one class** = **pure** → impurity = 0
* A node with **multiple classes mixed** → impurity increases

Decision Trees use these impurity scores to decide **which feature split is best**.

---



### **A. Gini Index**

**Definition:**
Gini measures how often a randomly chosen record would be **incorrectly classified** if we randomly assigned class labels based on class distribution.

* Low Gini = purer node
* High Gini = mixed node

**Intuition:**
“If I randomly pick two samples from the node, how likely are they from different classes?”

---

### **B. Entropy**

**Definition:**
Entropy measures the **disorder or randomness** in the node.

* High entropy = highly mixed (uncertain)
* Low entropy = more pure

**Intuition:**
“How unpredictable is the class label in this node?”

Inspired by **information theory** (Shannon entropy).

---

### **C. Information Gain (IG)**

**Definition:**
Information Gain tells us **how much impurity is reduced** by splitting on a particular feature.

[
IG = \text{Impurity(parent)} - \text{Weighted Impurity(children)}
]

A split is “good” when it **maximizes IG** (i.e., reduces impurity the most).

**IG is the criterion used by ID3/C4.5 Decision Tree algorithms.**

---

#### **3. Formula Comparison Table**

| Measure              | Formula                                 | Value Range       | Impurity Interpretation          | Decision Tree Use            |
| -------------------- | --------------------------------------- | ----------------- | -------------------------------- | ---------------------------- |
| **Gini Index**       | ( G = 1 - \sum p_i^2 )                  | 0 to 0.5 (binary) | Probability of misclassification | CART algorithm uses it       |
| **Entropy**          | ( H = -\sum p_i \log_2(p_i) )           | 0 to 1 (binary)   | Measures disorder & uncertainty  | ID3 / C4.5 algorithms use it |
| **Information Gain** | ( IG = H_{parent} - (w_LH_L + w_RH_R) ) | 0 to 1            | How much impurity decreases      | Used to select best split    |

---

### **4. Example to Understand Differences**

Suppose a node has 100 samples:

* 70 = Class A
* 30 = Class B

### **Compute Gini**

[
G = 1 - (0.7^2 + 0.3^2)
= 1 - (0.49 + 0.09)
= 0.42
]

### **Compute Entropy**

[
H = -(0.7 \log_2 0.7 + 0.3 \log_2 0.3)
\approx 0.881
]

### Observations:

* Gini is **always slightly lower** than Entropy
* Both indicate impurity
* Both rank splits very similarly in practice

---

### **5. Why Are These Measures Relevant in Decision Trees?**

### **Decision Trees choose the best feature splits using these impurity scores.**

### **At each node:**

1. Try possible splits (e.g., Age < 30, Income > 50k)
2. For each split:

   * Calculate Gini or Entropy for child nodes
   * Calculate weighted impurity
   * Compute Information Gain
3. Select the split that:

   * Minimizes impurity
   * Maximizes Information Gain

### **Outcome:**

* The tree becomes more structured
* Homogeneous leaves
* Better classification accuracy

---

#### Practical Differences in Decision Trees**

| Property            | Gini Index                                 | Entropy                              | Information Gain                  |
| ------------------- | ------------------------------------------ | ------------------------------------ | --------------------------------- |
| Calculation Speed   | Faster                                     | Slower (log operations)              | Derived from Entropy              |
| Sensitivity         | Slightly more sensitive to class imbalance | Strongly sensitive to minority class | Captures gain from split          |
| Preferred Algorithm | CART                                       | ID3, C4.5                            | Included in Entropy-based methods |
| Split Quality       | Very similar to Entropy                    | Very similar to Gini                 | Used for best split selection     |

**In practice:**
➡Gini and Entropy almost always choose the **same split**
➡Gini is faster
➡Entropy gives more weight to rare classes

---

### **7. Final Summary**

### **Gini Index**

* Measures impurity
* Lower is better
* Fast computation
* Used in CART Decision Trees

### **Entropy**

* Measures uncertainty
* Higher entropy = more mixed
* Used in ID3/C4.5
* Slightly slower

### **Information Gain**

* Measures the **drop in impurity** after splitting
* Used to choose the **best feature split**
* Higher IG = better split

## Math Notebook – Decision Tree Impurity Measures
This notebook contains pure math explanations and formulas.

### Entropy Formula
For a classification node with classes and probabilities:
\[
Entropy = -\sum p_i \log_2(p_i)
\]
Where:
- \( p_i \) is probability of class i

```python
import numpy as np
from math import log2

def entropy(probs):
    return -sum([p*log2(p) for p in probs if p>0])

entropy([0.5, 0.5])  # example
```

### Gini Impurity Formula
\[
Gini = 1 - \sum p_i^2
\]

```python
def gini(probs):
    return 1 - sum([p*p for p in probs])

gini([0.6, 0.4])  # example
```

### Information Gain
\[
IG = Impurity(parent) - (w_L * Impurity(left) + w_R * Impurity(right))
\]

```python
def information_gain(parent_imp, left_imp, right_imp, n_left, n_right):
    n = n_left + n_right
    return parent_imp - ((n_left/n)*left_imp + (n_right/n)*right_imp)
```

### Example Calculation

```python
parent = entropy([0.5, 0.5])
left = entropy([0.7, 0.3])
right = entropy([0.2, 0.8])
information_gain(parent, left, right, 40, 60)
```
