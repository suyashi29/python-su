### 1. What is Generative AI?

**Generative AI** refers to AI systems that can **create new content** — text, images, code, audio, video — based on patterns learned from existing data.

#### Traditional AI vs Generative AI

| Feature | Traditional AI | Generative AI |
|---------|---------------|---------------|
| **Task** | Classification, Prediction | Content Generation |
| **Output** | Labels, Numbers | Text, Images, Code |
| **Examples** | Spam Filter, Fraud Detection | ChatGPT, DALL-E, Copilot |
| **Approach** | Rule-based / Supervised | Deep Learning / Transformers |

**Key Insight:** Traditional AI **analyzes** data. Generative AI **creates** new data.

#### Simple Demo: Text Generation Using Markov Chains

A **Markov Chain** is one of the simplest forms of text generation — it predicts the next word based on the current word.

````python
import random

# --- Simple Markov Chain Text Generator ---

def build_markov_chain(text):
    """Build a simple word-level Markov chain from input text."""
    words = text.split()
    chain = {}
    for i in range(len(words) - 1):
        current_word = words[i]
        next_word = words[i + 1]
        if current_word not in chain:
            chain[current_word] = []
        chain[current_word].append(next_word)
    return chain

def generate_text(chain, start_word, length=15):
    """Generate text using the Markov chain."""
    word = start_word
    result = [word]
    for _ in range(length - 1):
        if word in chain:
            word = random.choice(chain[word])
            result.append(word)
        else:
            break
    return ' '.join(result)

# Training text (a small corpus)
corpus = """
Artificial intelligence is transforming the way businesses operate.
AI is helping companies automate tasks and improve efficiency.
Generative AI is creating new content like text images and code.
AI is the future of technology and innovation.
Businesses are using AI to gain competitive advantage.
"""

# Build the chain and generate text
chain = build_markov_chain(corpus)
print("Markov Chain Dictionary (partial):")
for key in list(chain.keys())[:5]:
    print(f"  '{key}' -> {chain[key]}")

print("\nGenerated Text:")
for i in range(3):
    print(f"  Attempt {i+1}: {generate_text(chain, 'AI', length=12)}")
````

### 3. Tokenization - How Machines Read Text

Tokenization is the **first step** in any NLP pipeline. It converts human-readable text into tokens that models can process.

**Types of Tokenization:**

- **Word-level:** `"Hello world"` becomes `["Hello", "world"]`
- **Character-level:** `"Hello"` becomes `["H", "e", "l", "l", "o"]`
- **Subword-level (BPE):** `"unhappiness"` becomes `["un", "happiness"]` (used by GPT)

````python
# --- Tokenization Demo ---

def word_tokenize(text):
    """Simple word-level tokenizer."""
    import re
    tokens = re.findall(r'\b\w+\b', text.lower())
    return tokens

def char_tokenize(text):
    """Character-level tokenizer."""
    return list(text)

def subword_tokenize(text):
    """Simulated subword tokenizer (simplified BPE-like)."""
    subwords = ['un', 'happi', 'ness', 'trans', 'form', 'er', 'gen', 'erat', 'ive',
                'art', 'ific', 'ial', 'intell', 'ig', 'ence']
    text_lower = text.lower()
    tokens = []
    i = 0
    while i < len(text_lower):
        matched = False
        for length in range(6, 1, -1):
            chunk = text_lower[i:i+length]
            if chunk in subwords:
                tokens.append(chunk)
                i += length
                matched = True
                break
        if not matched:
            if text_lower[i].isalnum():
                tokens.append(text_lower[i])
            i += 1
    return tokens

# Test with sample text
sample = "Generative Artificial Intelligence"

print(f"Original Text: '{sample}'")
print(f"\nWord Tokens:     {word_tokenize(sample)}")
print(f"Char Tokens:     {char_tokenize(sample)}")
print(f"Subword Tokens:  {subword_tokenize(sample)}")

print("\nToken Count Comparison:")
print(f"  Word-level:    {len(word_tokenize(sample))} tokens")
print(f"  Char-level:    {len(char_tokenize(sample))} tokens")
print(f"  Subword-level: {len(subword_tokenize(sample))} tokens")
````

### 4. Temperature and Sampling - Controlling Creativity

When LLMs generate text, they predict probabilities for the next token. **Temperature** controls how creative or deterministic the output is.

| Temperature | Behavior | Use Case |
|------------|----------|----------|
| 0.0 - 0.3 | Very focused, deterministic | Factual Q&A, code generation |
| 0.4 - 0.7 | Balanced | General conversation |
| 0.8 - 1.5 | Creative, diverse | Brainstorming, storytelling |

````python
import math
import random

# --- Temperature Sampling Demo ---

def apply_temperature(probabilities, temperature):
    """Apply temperature scaling to a probability distribution."""
    if temperature == 0:
        result = [0.0] * len(probabilities)
        result[probabilities.index(max(probabilities))] = 1.0
        return result
    
    logits = [math.log(p + 1e-10) / temperature for p in probabilities]
    max_logit = max(logits)
    exp_logits = [math.exp(l - max_logit) for l in logits]
    total = sum(exp_logits)
    return [e / total for e in exp_logits]

def sample_next_word(words, probabilities, temperature, num_samples=5):
    """Sample next words based on temperature-adjusted probabilities."""
    adjusted_probs = apply_temperature(probabilities, temperature)
    samples = random.choices(words, weights=adjusted_probs, k=num_samples)
    return samples, adjusted_probs

# Simulated next-word predictions
words = ["innovation", "technology", "strategy", "pizza", "unicorn"]
base_probs = [0.40, 0.30, 0.20, 0.07, 0.03]

print("Next word prediction after: 'The company focuses on...'")
print(f"Candidate words: {words}")
print(f"Base probabilities: {base_probs}")
print("=" * 60)

for temp in [0.1, 0.5, 1.0, 1.5]:
    samples, adj_probs = sample_next_word(words, base_probs, temp)
    print(f"\nTemperature = {temp}")
    print(f"  Adjusted probs: {[round(p, 3) for p in adj_probs]}")
    print(f"  5 Samples: {samples}")
````

### 6. Use Case 1: Simple Rule-Based Chatbot

A basic chatbot that responds to user queries using **pattern matching**. In production, this would be powered by an LLM like GPT-4 via Azure OpenAI.

````python
# --- Simple Enterprise Chatbot ---

import re

class SimpleEnterpriseBot:
    """A rule-based chatbot for enterprise FAQ."""
    
    def __init__(self):
        self.knowledge_base = {
            r'(hi|hello|hey)': 'Hello! I\'m your Enterprise AI Assistant. How can I help you today?',
            r'(leave|vacation|pto|time off)': 'You can apply for leave through the HR portal. You have 24 days of annual leave.',
            r'(salary|pay|payslip|compensation)': 'Salary slips are available on the HR portal under "My Payroll" section.',
            r'(password|reset|login)': 'To reset your password, visit the IT Self-Service portal or contact IT helpdesk at ext. 1234.',
            r'(training|learn|course|upskill)': 'Check out the Learning Management System (LMS) for available courses on GenAI, Cloud, and more!',
            r'(policy|policies|handbook)': 'Company policies are available on the intranet under "HR Policies" section.',
            r'(expense|reimbursement|claim)': 'Submit expense claims through the Finance portal within 30 days of the expense.',
            r'(bye|goodbye|exit|quit)': 'Goodbye! Have a great day!',
        }
        self.default_response = 'I\'m not sure about that. Please contact HR or IT support for assistance.'
    
    def respond(self, user_input):
        """Match user input against known patterns."""
        user_input_lower = user_input.lower().strip()
        for pattern, response in self.knowledge_base.items():
            if re.search(pattern, user_input_lower):
                return response
        return self.default_response

# Create bot and test with sample queries
bot = SimpleEnterpriseBot()

test_queries = [
    "Hi there!",
    "How do I apply for leave?",
    "Where can I find my salary slip?",
    "I need to reset my password",
    "Are there any AI training courses?",
    "What is the capital of France?",
    "Goodbye!"
]

print("Enterprise Chatbot Demo")
print("=" * 50)
for query in test_queries:
    print(f"\nUser: {query}")
    print(f"Bot:  {bot.respond(query)}")
````

### 7. Use Case 2: Text Summarization

A simple **extractive summarizer** that picks the most important sentences based on word frequency scoring.

````python
# --- Simple Extractive Text Summarizer ---

import re
from collections import Counter

def summarize_text(text, num_sentences=3):
    """
    Simple extractive summarizer using word frequency scoring.
    Steps:
    1. Tokenize into sentences and words
    2. Calculate word frequencies
    3. Score each sentence based on word frequencies
    4. Return top-scoring sentences
    """
    # Step 1: Split into sentences
    sentences = re.split(r'(?<=[.!?])\s+', text.strip())
    
    # Step 2: Calculate word frequencies (excluding common stop words)
    stop_words = {'the', 'a', 'an', 'is', 'are', 'was', 'were', 'in', 'on', 'at', 'to',
                  'for', 'of', 'and', 'or', 'but', 'with', 'has', 'have', 'had', 'be',
                  'been', 'being', 'it', 'its', 'this', 'that', 'these', 'those', 'by',
                  'from', 'as', 'can', 'will', 'do', 'does', 'did', 'not', 'no', 'so',
                  'if', 'then', 'than', 'too', 'very', 'just', 'about', 'also', 'more'}
    
    words = re.findall(r'\b\w+\b', text.lower())
    filtered_words = [w for w in words if w not in stop_words and len(w) > 2]
    word_freq = Counter(filtered_words)
    
    # Step 3: Score sentences
    sentence_scores = []
    for i, sentence in enumerate(sentences):
        score = sum(word_freq.get(w, 0) for w in re.findall(r'\b\w+\b', sentence.lower()))
        sentence_scores.append((score, i, sentence))
    
    # Step 4: Get top sentences (maintain original order)
    top_sentences = sorted(sentence_scores, key=lambda x: x[0], reverse=True)[:num_sentences]
    top_sentences = sorted(top_sentences, key=lambda x: x[1])
    
    summary = ' '.join([s[2] for s in top_sentences])
    return summary

# Sample enterprise document
document = """
Artificial Intelligence is rapidly transforming the enterprise landscape. Companies across 
industries are leveraging AI to automate repetitive tasks, enhance customer experiences, 
and drive data-driven decision making. Generative AI, powered by Large Language Models, 
represents the latest evolution in this transformation. These models can generate human-like 
text, summarize documents, write code, and even create marketing content. The adoption of 
GenAI in enterprises has grown by 250% in the past year alone. Organizations are investing 
heavily in AI infrastructure, talent acquisition, and responsible AI governance. However, 
challenges remain including data privacy concerns, model hallucinations, and the need for 
human oversight. Enterprises must develop clear AI strategies that balance innovation with 
risk management. Training employees on AI tools and building AI literacy across the 
organization is critical for successful adoption. The future of enterprise AI lies in 
combining human expertise with AI capabilities to create unprecedented value.
"""

print("Original Document:")
print(f"  Words: {len(document.split())} | Sentences: {len(re.split(r'(?<=[.!?])s+', document.strip()))}")
print("\n" + "=" * 60)
print("\nSummary (Top 3 Sentences):")
print(summarize_text(document, num_sentences=3))
print("\n" + "=" * 60)
print("\nSummary (Top 2 Sentences):")
print(summarize_text(document, num_sentences=2))
````

### 8. Use Case 3: Sentiment Analysis

A keyword-based **sentiment analyzer** that classifies text as Positive, Negative, or Neutral. In production, this would use fine-tuned models or Azure AI services.

````python
# --- Simple Sentiment Analyzer ---

class SentimentAnalyzer:
    """Keyword-based sentiment analyzer for demonstration."""
    
    def __init__(self):
        self.positive_words = {
            'good', 'great', 'excellent', 'amazing', 'wonderful', 'fantastic',
            'love', 'happy', 'satisfied', 'awesome', 'brilliant', 'outstanding',
            'perfect', 'best', 'impressive', 'innovative', 'efficient', 'helpful',
            'reliable', 'recommend', 'pleased', 'delighted', 'enjoy', 'success',
            'improved', 'exceptional', 'superb', 'positive', 'thrilled', 'valuable'
        }
        self.negative_words = {
            'bad', 'terrible', 'awful', 'horrible', 'poor', 'worst', 'hate',
            'disappointed', 'frustrating', 'waste', 'useless', 'broken',
            'slow', 'expensive', 'complicated', 'confusing', 'annoying',
            'unreliable', 'failure', 'problem', 'issue', 'difficult', 'lacking',
            'mediocre', 'subpar', 'unsatisfied', 'regret', 'unhappy', 'negative'
        }
        self.negation_words = {
            'not', 'no', 'never', 'neither', 'nobody', 'nothing',
            "don't", "doesn't", "isn't", "aren't", "wasn't", "weren't",
            "won't", "wouldn't", "shouldn't", "couldn't"
        }
    
    def analyze(self, text):
        """Analyze sentiment of given text."""
        words = text.lower().split()
        
        pos_score = 0
        neg_score = 0
        negate = False
        
        for word in words:
            clean_word = ''.join(c for c in word if c.isalnum() or c == "'")
            
            if clean_word in self.negation_words:
                negate = True
                continue
            
            if clean_word in self.positive_words:
                if negate:
                    neg_score += 1
                else:
                    pos_score += 1
                negate = False
            elif clean_word in self.negative_words:
                if negate:
                    pos_score += 1
                else:
                    neg_score += 1
                negate = False
            else:
                negate = False
        
        total = pos_score + neg_score
        if total == 0:
            return "Neutral", 0.0, pos_score, neg_score
        
        sentiment_score = (pos_score - neg_score) / total
        
        if sentiment_score > 0.2:
            label = "Positive"
        elif sentiment_score < -0.2:
            label = "Negative"
        else:
            label = "Neutral"
        
        return label, round(sentiment_score, 2), pos_score, neg_score

# Create analyzer
analyzer = SentimentAnalyzer()

# Test with various enterprise-related reviews
reviews = [
    "The new AI tool is amazing and has greatly improved our team's efficiency!",
    "This software is terrible. It's slow, confusing, and completely unreliable.",
    "The product is okay. It has some good features but also some issues.",
    "I love the innovative approach. The results are outstanding and impressive!",
    "The service is not good. Customer support was disappointing and unhelpful.",
    "The training session was excellent and very helpful for our team.",
]

print("Sentiment Analysis Results")
print("=" * 70)
for review in reviews:
    label, score, pos, neg = analyzer.analyze(review)
    print(f"\nReview: \"{review}\"")
    print(f"  Sentiment: {label} | Score: {score} | (Positive: {pos} / Negative: {neg})")
````

### 9. Use Case 4: Email Drafting Assistant

A template-based **email generator** for common enterprise scenarios. In production, LLMs generate fully customized, context-aware emails.

````python
# --- Email Drafting Assistant ---

from datetime import datetime

class EmailDraftAssistant:
    """Template-based email drafting assistant."""
    
    def __init__(self):
        self.templates = {
            "meeting_invite": {
                "subject": "Meeting Invitation: {topic}",
                "body": """Dear {recipient},

I hope this message finds you well.

I would like to invite you to a meeting to discuss {topic}.

Date: {date}
Time: {time}
Location: {location}

Agenda:
{agenda}

Please confirm your availability at your earliest convenience.

Best regards,
{sender}"""
            },
            "project_update": {
                "subject": "Project Update: {project_name}",
                "body": """Dear {recipient},

I am writing to provide an update on the {project_name} project.

Status: {status}
Progress: {progress}%
Next Milestone: {milestone}

Key Highlights:
{highlights}

Please let me know if you have any questions or concerns.

Best regards,
{sender}"""
            },
            "follow_up": {
                "subject": "Follow-Up: {topic}",
                "body": """Dear {recipient},

I hope you are doing well. I wanted to follow up on our recent discussion regarding {topic}.

{follow_up_details}

Looking forward to hearing from you.

Best regards,
{sender}"""
            }
        }
    
    def draft_email(self, template_type, **kwargs):
        """Generate an email draft from a template."""
        if template_type not in self.templates:
            return "Template not found. Available: " + ", ".join(self.templates.keys())
        
        template = self.templates[template_type]
        subject = template["subject"].format(**kwargs)
        body = template["body"].format(**kwargs)
        
        return f"Subject: {subject}\n{'='*50}\n{body}"

# Create assistant
email_bot = EmailDraftAssistant()

# Generate a meeting invite
print(email_bot.draft_email(
    "meeting_invite",
    recipient="Team",
    topic="GenAI Training - Day 2",
    date="2026-05-06",
    time="10:00 AM IST",
    location="Virtual (Microsoft Teams)",
    agenda="- Recap of Day 1\n- Hands-on with Azure OpenAI\n- Building a RAG Pipeline\n- Q&A Session",
    sender="Suyashi"
))

print("\n" + "-" * 60 + "\n")

# Generate a project update
print(email_bot.draft_email(
    "project_update",
    recipient="Leadership Team",
    project_name="AI-Powered Customer Service Bot",
    status="On Track",
    progress="75",
    milestone="UAT Testing - May 15, 2026",
    highlights="- Chatbot accuracy improved to 92%\n- Integration with CRM completed\n- User training scheduled for next week",
    sender="Suyashi"
))
````
