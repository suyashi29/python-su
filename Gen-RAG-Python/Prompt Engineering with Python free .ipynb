# 1. What is Prompt Engineering?

Prompt Engineering is the process of designing effective instructions (prompts) for Large Language Models (LLMs) to generate accurate, structured, and useful outputs.

A good prompt helps:
- Improve response quality
- Reduce hallucinations
- Control output format
- Guide reasoning
- Improve RAG applications
- Automate enterprise workflows

Examples:
- Text summarization
- Document Q&A
- Code generation
- Chatbots
- Data extraction


# 2. Basic Prompt Structure

A good prompt generally contains:

## Components
1. Role
2. Task
3. Context
4. Constraints
5. Output Format

## Example

```text
You are a data analyst.
Analyze the sales data for Q1.
Return only 5 bullet points.
Use simple business language.
```

# 4.1 Zero-shot Prompting

The model is asked to perform a task without examples.


````python
prompt = '''
Classify the sentiment of the following review:

"The product quality is amazing and delivery was very fast."
'''

print(prompt)
````

## Expected Output
Positive Sentiment


# 4.2 One-shot Prompting

The model is given one example before the actual task.


````python
prompt = '''
Example:
Text: "The movie was fantastic."
Sentiment: Positive

Now classify:
Text: "The service was disappointing."
'''

print(prompt)
````

# 4.3 Few-shot Prompting

Multiple examples are provided to improve accuracy.


````python
prompt = '''
Example 1:
Question: What is AI?
Answer: Artificial Intelligence.

Example 2:
Question: What is ML?
Answer: Machine Learning.

Now answer:
Question: What is NLP?
'''

print(prompt)
````

# 4.4 Chain-of-Thought Prompting

The model is encouraged to think step-by-step.


````python
prompt = '''
A company sold 120 laptops in January and 150 in February.

Calculate:
1. Total laptops sold
2. Percentage increase

Think step-by-step.
'''

print(prompt)
````

# 4.5 Role-based Prompting

Assigning a role helps control the response style.


````python
prompt = '''
You are a cloud architect.

Explain Kubernetes to a beginner using simple examples.
'''

print(prompt)
````

# 4.6 Instruction-based Prompting

Clear instructions improve output consistency.


````python
prompt = '''
Summarize the following paragraph in exactly 3 bullet points.

Paragraph:
Generative AI is transforming enterprise workflows by automating repetitive tasks,
improving productivity, and enabling intelligent assistants.
'''

print(prompt)
````

# 4.7 Structured Output Prompting

Useful for APIs, automation, and RAG systems.


````python
prompt = '''
Extract the following details in JSON format:

Name: Ashi
Project: RAG Modernization
Technology: LangChain

Return valid JSON only.
'''

print(prompt)
````

# 4.8 Contextual Prompting for RAG

RAG systems use external knowledge as context.


````python
context = '''
Adobe Experience Platform helps organizations manage customer experiences.
'''

question = "What does Adobe Experience Platform help with?"

prompt = f'''
Use the context below to answer the question.

Context:
{context}

Question:
{question}
'''

print(prompt)
````

# 5. Prompt Engineering Best Practices

## Tips
- Be specific
- Define output format
- Use delimiters
- Provide examples
- Assign roles
- Keep prompts modular
- Avoid ambiguity
- Use context effectively in RAG systems

## Common Problems
| Problem | Solution |
|---|---|
| Hallucination | Add context |
| Long answers | Add limits |
| Incorrect format | Specify JSON/table |
| Weak reasoning | Use Chain-of-Thought |

````python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="distilgpt2"
)

prompt = '''
Explain life in simple terms.
'''

result = generator(
    prompt,
    max_length=80,
    num_return_sequences=1
)

print(result[0]['generated_text'])
````
