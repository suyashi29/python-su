## What is EDA (Exploratory Data Analysis)?

EDA is the process of understanding your dataset before modeling by:

- Summarizing data (mean, median, distribution)
- Detecting patterns & relationships
- Identifying anomalies/outliers
- Validating assumptions

## In your case:
**Goal = Identify factors influencing hiring (Hired)**

### Step 1: Understand the Dataset

#### Columns:

- YearsExperience → numeric
- EducationLevel → categorical
- SkillsScore → numeric
- CertificationCount → numeric
- Hired → target (0/1)

### Business Question: What increases the probability of getting hired?

### Step 2: Data Cleaning
- Check missing values
- Check duplicates
- Correct data types

### Step 3: Univariate Analysis: Analyze each feature individually:
- Distribution of experience
- Education level frequency
- Skills score spread
- Hiring ratio

### Step 4: Bivariate Analysis : Compare each feature with Hired

***Examples***

- Experience vs Hiring
- Education vs Hiring
- SkillsScore vs Hiring
- Certification vs Hiring

### Step 5: Correlation Analysis: Numerical relationships:

- Which variables are strongly linked with hiring?

### Step 6: Feature Engineering (VERY IMPORTANT)

- Create new insights:
- Experience Buckets (Junior/Mid/Senior)
- Skills per Certification
- Education Encoding

### Step 7: Insights & Conclusion

- Answer:What factors most influence hiring?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import gradio as gr
```

```python
# Load Data
def load_data():
    #data = pd.read_csv("hiredata.csv")
    data = pd.read_csv(r"https://raw.githubusercontent.com/suyashi29/python-su/refs/heads/master/Applied%20Data%20Modelling%20using%20Gradio/hiredata.csv")
    return data
```

```python
data= pd.read_csv(r"https://raw.githubusercontent.com/suyashi29/python-su/refs/heads/master/Applied%20Data%20Modelling%20using%20Gradio/hiredata.csv")

data.head(3) # tail()
```

```python
# Step 1: Overview
def data_overview():
    df = load_data()
    return df.describe(include="all").to_string()
```

```python
# Step 2: Univariate Analysis
def plot_univariate(column):
    df = load_data()
    plt.figure()
    sns.histplot(df[column], kde=True)
    plt.title(f"Distribution of {column}")
    return plt

# Step 3: Bivariate Analysis
def plot_bivariate(column):
    df = load_data()
    plt.figure()
    sns.boxplot(x="Hired", y=column, data=df)
    plt.title(f"{column} vs Hiring")
    return plt

# Step 4: Correlation
def correlation_plot():
    df = load_data()
    plt.figure()
    sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
    return plt

# Step 5: Feature Engineering
def feature_engineering():
    df = load_data()
    
    # Experience Bucket
    df["ExperienceLevel"] = pd.cut(
        df["YearsExperience"],
        bins=[0,5,10,20],
        labels=["Junior", "Mid", "Senior"]
    )
    
    # Skills per Certification
    df["SkillPerCert"] = df["SkillsScore"] / (df["CertificationCount"] + 1)
    
    return df.to_string()

# Step 6: Final Insights
def insights():
    return """
Key Observations:
1. Higher experience → Higher hiring probability
2. PhD/Post-Graduate candidates more likely hired
3. SkillsScore positively impacts hiring
4. Certifications help but not as strong as skills
"""

```

```python
with gr.Blocks() as app:
    gr.Markdown("#  Hiring Dashboard Analysis")

    # Output components
    summary_output = gr.Textbox(label="Summary")
    uni_plot = gr.Plot()
    bi_plot = gr.Plot()
    corr_plot = gr.Plot()
    feature_output = gr.Textbox()
    insight_output = gr.Textbox()

    gr.Markdown("## Data Overview")
    gr.Button("Show Summary").click(fn=data_overview, outputs=summary_output)

    gr.Markdown("## Univariate Analysis")
    col1 = gr.Dropdown(["YearsExperience", "SkillsScore", "CertificationCount"])
    gr.Button("Plot").click(fn=plot_univariate, inputs=col1, outputs=uni_plot)

    gr.Markdown("## Bivariate Analysis")
    col2 = gr.Dropdown(["YearsExperience", "SkillsScore", "CertificationCount"])
    gr.Button("Compare with Hiring").click(fn=plot_bivariate, inputs=col2, outputs=bi_plot)

    gr.Markdown("## Correlation Heatmap")
    gr.Button("Show Correlation").click(fn=correlation_plot, outputs=corr_plot)

    gr.Markdown("## Feature Engineering")
    gr.Button("Generate Features").click(fn=feature_engineering, outputs=feature_output)

    gr.Markdown("## Final Insights")
    gr.Button("Show Insights").click(fn=insights, outputs=insight_output)

app.launch()
```
