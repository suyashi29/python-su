Gradio is a Python library used to quickly build interactive web UIs for ML models or functions. It wraps your Python function into a browser-based app with inputs and outputs.

A Gradio app =
**Function (logic) + Inputs + Outputs + Interface wrapper**

### Step-by-Step Implementation
#### 1. Import a install Gradio App 

```python
### pip install gradio
import gradio as gr
```

### 2. Define a Function

This is your core logic.

Example: Simple Addition

```python
def add_numbers(a, b):
    return a + b
```

### 3. Create Interface 

```python
interface = gr.Interface(
    fn=add_numbers,        # Function
    inputs=["number", "number"],  # Input types
    outputs="number"       # Output type
)
```

### 4. Launch App 

```python
interface.launch()
```

### Key Parameters of gr.Interface
- fn (Function)
The backend logic
Must accept inputs → return outputs

### Inputs : Defines UI components:

| Type     | Example       |
| -------- | ------------- |
| "text"   | Text input    |
| "number" | Numeric input |
| "slider" | Range input   |
| "image"  | Image upload  |


```python
inputs=["text"]
```

### outputs: Defines output display
| Type     | Example        |
| -------- | -------------- |
| "text"   | Text result    |
| "number" | Numeric output |
| "label"  | Classification |
| "image"  | Image output   |


### title & description

```python
gr.Interface(
    fn=add_numbers,
    inputs=["number", "number"],
    outputs="number",
    title="Addition App",
    description="Enter two numbers"
)
```

### Example: ML Model (Churn Prediction)

```python
def predict_churn(age, purchase_freq, avg_spend):
    score = age * 0.1 + purchase_freq * -0.5 + avg_spend * 0.01
    return "Churn" if score > 5 else "No Churn"

app = gr.Interface(
    fn=predict_churn,
    inputs=["number", "number", "number"],
    outputs="text",
    title="Customer Churn Predictor"
)

app.launch()
```

### Using gr.Blocks (Advanced UI)

```python
with gr.Blocks() as demo:
    num1 = gr.Number()
    num2 = gr.Number()
    output = gr.Number()

    btn = gr.Button("Add")
    btn.click(fn=add_numbers, inputs=[num1, num2], outputs=output)

demo.launch()
```
