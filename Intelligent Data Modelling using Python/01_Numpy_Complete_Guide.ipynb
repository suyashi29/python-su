# NumPy Complete Guide

## What is NumPy?

NumPy (Numerical Python) is the fundamental library for numerical computing in Python. It provides fast multidimensional arrays and a rich collection of mathematical, statistical, linear algebra, and random number generation functions.

### Why NumPy?
- Faster than Python lists
- Memory efficient
- Supports vectorized operations
- Foundation for Pandas, Scikit-learn, TensorFlow, and PyTorch

In this notebook, you'll learn the major NumPy functions with examples and practice exercises.


```
import numpy as np
print('NumPy Version:', np.__version__)
```

## Creating Arrays

```
a=np.array([1,2,3,4])
b=np.arange(1,11)
c=np.linspace(0,100,11)
d=np.zeros((2,3))
e=np.ones((2,3))
f=np.eye(3)
print(a,b,c,d,e,f,sep='\n\n')
```

## Array Attributes

```
arr=np.arange(12).reshape(3,4)
print(arr)
print("Shape:",arr.shape)
print("Dimensions:",arr.ndim)
print("Size:",arr.size)
print("Datatype:",arr.dtype)
print("Item Size:",arr.itemsize)
```

## Indexing and Slicing

```
print(arr[1,2])
print(arr[:,1])
print(arr[1:])
print(arr[::-1])
```

## Reshaping

```
x=np.arange(12)
print(x.reshape(3,4))
print(x.flatten())
print(x.reshape(3,4).T)
```

## Mathematical Functions

```
print(np.sqrt(np.arange(1,10)))
print(np.power(np.arange(1,6),2))
print(np.exp([1,2,3]))
print(np.log([1,2,3]))
```

## Statistical Functions

```
marks=np.array([75,82,90,68,99,55,88,72])
print(np.mean(marks))
print(np.median(marks))
print(np.std(marks))
print(np.min(marks))
print(np.max(marks))
print(np.sum(marks))
```

## Sorting and Searching

```
arr=np.array([5,8,1,7,3])
print(np.sort(arr))
print(np.argsort(arr))
print(np.where(arr>4))
print(np.unique([1,2,2,3,4,4]))
```

## Boolean Masking

```
arr=np.arange(1,21)
print(arr[arr%2==0])
print(arr[arr>10])
```

## Broadcasting

```
a=np.array([1,2,3])
print(a+10)
print(a*5)
```

## Random Module

```
np.random.seed(42)
print(np.random.rand(5))
print(np.random.randint(1,100,10))
```

## Linear Algebra

```
A=np.array([[1,2],[3,4]])
B=np.array([[5,6],[7,8]])
print(np.dot(A,B))
print(np.linalg.det(A))
print(np.linalg.inv(A))
```

## Useful Functions

```
a=np.array([1,2,3])
print(np.append(a,4))
print(np.insert(a,1,100))
print(np.delete(a,1))
print(np.repeat(a,2))
print(np.tile(a,3))
print(np.flip(a))
```
