Gradio is a Python library used to quickly build interactive web UIs for ML models or functions. It wraps your Python function into a browser-based app with inputs and outputs.

A Gradio app =
**Function (logic) + Inputs + Outputs + Interface wrapper**

### Step-by-Step Implementation
#### 1. Import a install Gradio App 

```python
pip install gradio
```

```python
### pip install gradio
import gradio as gr
```

### 2. Define a Function

This is your core logic.

Example: Simple Addition

```python
def add_numbers(a, b):
    return a + b
```

### 3. Create Interface 

```python
interface = gr.Interface(
    fn=add_numbers,        # Function
    inputs=["number", "number"],  # Input types
    outputs="number"       # Output type
)
```

### 4. Launch App 

```python
interface.launch()
```

### Example: ML Model (Churn Prediction)

```python
def predict_churn(age=0.0, purchase_freq=0, avg_spend=0):
    score = age * 0.1 + purchase_freq * -0.5 + avg_spend * 0.01
    return "Churn" if score > 5 else "No Churn"

app = gr.Interface(
    fn=predict_churn,
    inputs=["number", "number", "number"],
    outputs="text",
    title="Customer Churn Predictor"
)

app.launch()
```

### Gradio to solve Quadratic Equation using Gradio

```python
import gradio as gr
import math

def solve_and_evaluate(a, b, c, x):
    a, b, c, x = float(a), float(b), float(c), float(x)
    
    if a == 0:
        return "Not a quadratic equation (a cannot be 0)."
    
    # Discriminant
    discriminant = b**2 - 4*a*c
    
    # Roots calculation
    if discriminant > 0:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        roots = f"Two Real Roots: {root1}, {root2}"
    
    elif discriminant == 0:
        root = -b / (2*a)
        roots = f"One Real Root: {root}"
    
    else:
        real_part = -b / (2*a)
        imag_part = math.sqrt(-discriminant) / (2*a)
        roots = f"Complex Roots: {real_part} ± {imag_part}i"
    
    # Function evaluation
    fx = a*(x**2) + b*x + c
    
    return f"{roots}\n\nf({x}) = {fx}"

# Gradio Interface
iface = gr.Interface(
    fn=solve_and_evaluate,
    inputs=[
        gr.Number(label="Coefficient a"),
        gr.Number(label="Coefficient b"),
        gr.Number(label="Coefficient c"),
        gr.Number(label="Enter x value")
    ],
    outputs="text",
    title="Quadratic Solver + Evaluator",
    description="Solve ax² + bx + c = 0 and evaluate f(x)"
)

iface.launch()
```

```python
import gradio as gr

def calculator(a, b,p,operation):
    a, b,p = float(a), float(b),int(p)
    
    if operation == "Add":
        return a + b
    elif operation == "Subtract":
        return a - b
    elif operation == "Multiply":
        return a * b
    elif operation == "Divide":
        return "Cannot divide by zero" if b == 0 else a / b
    elif operation == "power":
        return a**p 

iface = gr.Interface(
    fn=calculator,
    inputs=[
        gr.Number(label="Enter first number"),
        gr.Number(label="Enter second number"),
        gr.Number(label="Enter Power for first number"),
        gr.Radio(["Add", "Subtract", "Multiply", "Divide","power"], label="Operation")
    ],
    outputs="text",
    title="Basic Calculator"
)

iface.launch()
```

## To add two arrays

```python
import gradio as gr
import numpy as np

def add_arrays(arr1, arr2):
    try:
        # Convert input (string → array)
        arr1 = np.array([float(x) for x in arr1.split(",")])
        arr2 = np.array([float(x) for x in arr2.split(",")])
        
        if arr1.shape != arr2.shape:
            return "Error: Arrays must have same shape"
        
        result = arr1 + arr2   # true array operation
        return result.tolist() # Convert back for display
    
    except Exception as e:
        return f"Error: {str(e)}"

iface = gr.Interface(
    fn=add_arrays,
    inputs=[
        gr.Textbox(label="Array 1 (e.g., 1,2,3)"),
        gr.Textbox(label="Array 2 (e.g., 4,5,6)")
    ],
    outputs="text",
    title="Array Addition using NumPy"
)

iface.launch()
```
