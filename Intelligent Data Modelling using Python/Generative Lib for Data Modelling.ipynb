### GenAI-powered tools 
Can build a machine learning model with minimal coding. These are excellent to demonstrate in a training session because learners can see a complete data modelling workflow in just a few minute
1. Pycaret
- Cleans missing values
- Encodes categorical variables
- Scales data (if appropriate)
- Trains multiple algorithms
- Compares performance
- Selects the best model
**This is one of the easiest ways to demonstrate end-to-end modelling**
2.  H2O AutoML

  It automatically trains and compares many algorithms such as Gradient Boosting, Random Forest, XGBoost, and Stacked Ensembles.
  
  | Tool                     | Best For                | Coding Required | Time to First Model |
| ------------------------ | ----------------------- | --------------- | ------------------- |
| PyCaret                  | AutoML in Python        | Low             | 5–10 min            |

```python
from pycaret.classification import *

hire = pd.read_csv("hiredata.csv")
setup(data=hire,
      target="Hired",
      session_id=123)

best_model = compare_models()

evaluate_model(best_model)

predict_model(best_model)
```
